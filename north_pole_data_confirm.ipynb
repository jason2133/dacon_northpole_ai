```
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
def sorted_list(path):
  tmplist = glob.glob(path)
  tmplist.sort()
  return tmplist

def read_csv(path):
  df = pd.read_csv(path)
  return df

def load_npy(path):
  npy = np.load(path)
  return npy

def show_data(npy):
  num_channel = npy.shape[-1]
  plt.figure(figsize=(3*num_channel, 5))
  for channel in range(num_channel):
    tmpimg = npy[:, :, channel]
    plt.subplot(1, num_channel, channel+1)
    plt.imshow(tmpimg)
  plt.tight_layout()
  plt.show()
  plt.close()
```
